```python
import splitfolders
input_folder = './data_deep_learning_iris_v3'
output_folder = './split_data_iris_v3'
 
#split the images into 80% training and 20% testing sets
splitfolders.ratio(input_folder, output_folder, seed=42, ratio=(0.8, 0.2))
```

```python
import keras
import numpy as np
import matplotlib.pyplot as plt
from keras.preprocessing.image import ImageDataGenerator

# Define the directories for the training and validation data
train_dir = './split_deep_learning_data_iris_v2/train'
validation_dir = './split_deep_learning_data_iris_v2/val'

# Define the parameters for the data augmentation
train_datagen = ImageDataGenerator(rescale=1./255,
                                   rotation_range=40,
                                   width_shift_range=0.2,
                                   height_shift_range=0.2,
                                   shear_range=0.2,
                                   zoom_range=0.2,
                                   horizontal_flip=True,
                                   fill_mode='nearest')

validation_datagen = ImageDataGenerator(rescale=1./255)

# Load the training and validation data
train_generator = train_datagen.flow_from_directory(train_dir,
                                                    target_size=(270, 1800),
                                                    batch_size=20,
                                                    class_mode='categorical')

validation_generator = validation_datagen.flow_from_directory(validation_dir,
                                                              target_size=(270, 1800),
                                                              batch_size=20,
                                                              class_mode='categorical')

# Define the ResNet model architecture
model = keras.applications.resnet50.ResNet50(weights=None, input_shape=(270, 1800, 3), classes=3)

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(
    train_generator, # the train generator
    steps_per_epoch=20, # the number of batches per epoch
    validation_data=validation_generator, # the test generator
    epochs=10 # the number of epochs
)

# Evaluate the model
test_loss, test_acc = model.evaluate(validation_generator, steps=50)
print('Test accuracy:', test_acc)
```

```python
import cv2
import pyautogui
import numpy as np
from keras.models import load_model
import keyboard
import mss.tools
import time
import uuid
import pyautogui
import tensorflow as tf

model = load_model('model_v2_iris.h5')
#model = load_model('model_v3_iris.h5')

def predict_direction(image):
    # Preprocess the image
    image = cv2.resize(image, (270, 1800))
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    #image = np.reshape(image, (1, -1))
    image = np.expand_dims(image, axis=0)

    # Predict the direction
    #direction = model.predict(image)[1][0][0]
    direction = model.predict(image)
    return direction

def screenshot():
    with mss.mss() as sct:
        time.sleep(1)
        monitor_info = sct.monitors[0]
        top_margin = (monitor_info["height"] // 4) + 50
        monitor = {"top": monitor_info["top"] + top_margin, "left": monitor_info["left"] + 55, "width": monitor_info["width"] - 120, "height": (monitor_info["height"] * 2 // 3)-450}
        output = f"test/{str(uuid.uuid4())}.png".format(**monitor)
        sct_img = sct.grab(monitor)
        mss.tools.to_png(sct_img.rgb, sct_img.size, output=output)
    return output     

while not keyboard.is_pressed("space"):
    img_path = screenshot()
    img = cv2.imread(img_path)
    frame = np.array(img)
    frame = cv2.cvtColor(frame, cv2.COLOR_RGB2BGR)
    dir = predict_direction(frame)
    predicted_class = np.argmax(dir)  # Get the index of the class with the highest probability
    print(predicted_class)
    threshold = 0.40

    confidence = dir[0][predicted_class]

    print(f"Predicted Index: {predicted_class}, Confidence: {confidence}")
    #if confidence < threshold:
    if (predicted_class == 0):
        keyboard.press('d')
        pyautogui.mouseDown()
        time.sleep(0.15)
        pyautogui.mouseUp()
        keyboard.release('d')
    elif (predicted_class == 2):
        keyboard.press('q')
        pyautogui.mouseDown()
        time.sleep(0.15)
        pyautogui.mouseUp()
        keyboard.release('q')
    else:
        pyautogui.mouseDown()             
        pyautogui.mouseUp()
```
