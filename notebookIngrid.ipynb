```python
import mss
with mss.mss() as sct:
    monitor = sct.monitors[1] # Use the first monitor
    width = monitor["width"] # Get the width of the screen
    height = monitor["height"] # Get the height of the screen

    top = height // 2 # The top pixel value is half of the height
    left = 0 # The left pixel value is zero
    right = width # The right pixel value is the same as the width
    bottom = height # The bottom pixel value is the same as the height
    # The screen part to capture
    #monitor = {"top": 0, "left": 0, "width": 1080, "height": 920}
    monitor = {"top": top, "left": left, "width": right, "height": bottom}
    output = "sct-{top}x{left}_{width}x{height}.png".format(**monitor)

    # Grab the data
    sct_img = sct.grab(monitor)

    # Save to the picture file
    mss.tools.to_png(sct_img.rgb, sct_img.size, output=output)
    print(output)
```

```python
import mss.tools
import os

# Specify the folder where you want to save the images
output_folder = "ingridimages"
time.sleep(20)
with mss.mss() as sct:
    monitor_info = sct.monitors[0]

    # The screen part to capture
    monitor = {"top": monitor_info["top"] + monitor_info["height"] // 2, "left": monitor_info["left"], "width": monitor_info["width"], "height": monitor_info["height"] // 2}

    # Run the loop 10 times
    for i in range(10):
        time.sleep(2)
        # Construct the full output path with the specified folder and loop variable i
        output = os.path.join(output_folder, "sct-{top}x{left}_{width}x{height}{i}.png".format(**monitor, i=i))

        # Grab the data
        sct_img = sct.grab(monitor)

        # Save to the picture file
        mss.tools.to_png(sct_img.rgb, sct_img.size, output=output)
        print(output)
```

```python
img_orig = cv2.imread('dataset/97639b03-caf6-4048-adad-f6ee77bb69c8.png')
img_rgb = cv2.cvtColor(img_orig, cv2.COLOR_BGR2RGB)

# canny detection without blurring
#img_canny = cv2.Canny(img_rgb, threshold1=321, threshold2=400)
img_canny = cv2.Canny(img_rgb, threshold1=280, threshold2=400)
plt.figure(figsize = (15, 15))
plt.subplot(1, 2, 1); plt.imshow(img_rgb)
plt.axis('off')
plt.subplot(1, 2, 2); plt.imshow(img_canny)
plt.axis('off')
```

```python
# set the lower and upper threshold
med_val = np.median(img_orig)
lower = int(max(0, .7 * med_val))
upper = int(min(255, 1.3 * med_val))

# blurring with ksize = 3
img_k3 = cv2.blur(img_orig, ksize = (3, 3))
# canny detection with different thresholds
edges_k3 = cv2.Canny(img_k3, threshold1 = lower, threshold2 = upper)
edges_k3_2 = cv2.Canny(img_k3, lower, upper + 75)

# blurring with ksize = 5 
img_k5 = cv2.blur(img_orig, ksize = (5, 5))
# canny detection with different thresholds
edges_k5 = cv2.Canny(img_k5, lower, upper)
edges_k5_2 = cv2.Canny(img_k5, lower, upper + 75)

# plot the images
images = [edges_k3, edges_k3_2, edges_k5, edges_k5_2]
titles = ['blurring, kernel = 3x3, lower threshold', 
          'blurring, kernel = 3x3, higher threshold', 
          'blurring, kernel = 5x5, lower threshold', 
          'blurring, kernel = 5x5, higher threshold']
plt.figure(figsize = (20, 15))
for i in range(4):
    plt.subplot(2, 2, i+1)
    plt.imshow(images[i])
    plt.axis('off')
    plt.title(titles[i])
        
plt.show()
```
